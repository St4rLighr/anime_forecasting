## Dealing with nulls

```python
anime_df.info()
```

```python
#checking for nulls
anime_df.isnull().sum()
```

```python
#dropping nulls in title column first

anime_df.dropna(subset="title",inplace=True)
```

```python
anime_df.dropna(subset="idMal",inplace=True) #contains join key
```

### Changing dtypes

```python
#int64 columns - idMal[1], episodes[4], duration[5],rankings_rank[23]

#date column - startDay[7], startMonth[8], startYear[9]

#float columns - meanScore,averageScore
```

```python
cols_to_format = [anime_df.columns[1]] + anime_df.columns[4:6].tolist() + anime_df.columns[7:10].tolist() + [anime_df.columns[23]]
for col in cols_to_format:
    anime_df[col] = anime_df[col].astype('Int64')
```

```python
anime_df.info()
```

### Filling nulls with Mode/Mean

```python
#columns to add mode - episodes (drop??) -- keep ongoing anime in new variable, delete if it's just blank
#columns to add avg - startDay, startMonth, startYear,rankings_rank
#dropping rankings columns (exclude: rankings_rank,rankings_context) -- has more than 50% that's empty, isDisabled
#replace nan rankings_context, sites and type with unknown
#replace date and month with 01
```

```python
anime_df.dropna(subset="startYear",inplace=True) #dropping NaN start year as it isnt much missing
```

```python
anime_df[['rankings_context', 'site','type']] = anime_df[['rankings_context', 'site','type']].fillna('Unknown') #unknown added to obj columns
```

```python
# Drop columns multiple column
anime_df = anime_df.drop(['averageScore','isDisabled', 'rankings_allTime','rankings_id', 'rankings_format','rankings_type', 'rankings_year'], axis=1)
```

```python
anime_df.isnull().sum()
```

```python
anime_df[['startDay', 'startMonth']] = anime_df[['startDay','startMonth']].fillna(value=1)
```

```python
anime_df.isnull().sum()
```

```python
anime_df.dropna(subset="duration", inplace=True) #dropping NaN duration
```

```python
anime_df.isnull().sum()
```

```python
import datetime
```

```python
# anime_df[['startDay', 'startMonth']] replace nulls with 01 -- fillna()
# drop null startyear
```

```python
anime_df['concate_date'] = anime_df[['startDay', 'startMonth','startYear']].astype(str).apply(lambda x: '-'.join(x), axis=1)
```

```python
anime_df['concate_date']
```

```python
anime_df['start_date'] = pd.to_datetime(anime_df['concate_date'],format='%d-%m-%Y')
```

```python
anime_df['start_date'].info()
```

```python
# Dropping the temporary concate_date column
anime_df = anime_df.drop(columns=['concate_date'])
```

```python
anime_df.isnull().sum()
```

```python
anime_df_drop_ranks = anime_df.drop('rankings_rank', axis=1)
```

```python
anime_df_drop_ranks.info()
```

```python
dropped_eps_rank = anime_df.dropna(subset="episodes") #removing rank column anime_df.drop(['rankings_rank'], axis=1)
```

```python
dropped_eps_rank = dropped_eps_rank.drop('rankings_rank', axis=1)

# dropped_eps_rank.drop(['rankings_rank'], axis=1,inplace=True)
```

```python
dropped_eps_rank.info()
```

### Expanding Columns - Genres and Tags

Available Dataframes:

anime_df

anime_df_drop_ranks

dropped_eps_rank

```python
anime_df.info() #columns to expand genres, tags
```

```python
def expand_column(column_name):
    return column_name.strip("[]").replace("'", "").split(", ")
```

```python
# Create a copy of the original dataframes
anime_df_copy = anime_df.copy()
anime_df_drop_ranks_copy = anime_df_drop_ranks.copy()
dropped_eps_rank_copy = dropped_eps_rank.copy()
```

```python
anime_df_copy['genres'] = anime_df_copy['genres'].apply(expand_column)
anime_df_copy['tags'] = anime_df_copy['tags'].apply(expand_column)
```

```python
anime_df_exploded = anime_df_copy.explode('genres').explode('tags')
```

```python
anime_df_drop_ranks_copy['genres'] = anime_df_drop_ranks_copy['genres'].apply(expand_column)
anime_df_drop_ranks_copy['tags'] = anime_df_drop_ranks_copy['tags'].apply(expand_column)
```

```python
anime_df_drop_ranks_exploded = anime_df_drop_ranks_copy.explode('genres').explode('tags')
```

```python
dropped_eps_rank_copy['genres'] = dropped_eps_rank_copy['genres'].apply(expand_column)
dropped_eps_rank_copy['tags'] = dropped_eps_rank_copy['tags'].apply(expand_column)
```

```python
dropped_eps_rank_exploded = dropped_eps_rank_copy.explode('genres').explode('tags')
```
