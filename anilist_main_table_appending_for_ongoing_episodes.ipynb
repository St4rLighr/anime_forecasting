### Table 3

*IGNORE NULLS IN EPISODES!!!!

To Do

drop rankings_rank,id_1, idMal_1, site, type columns

drop null sites

rename sites to external_sites

expand genres and tags columns

```python
table_3 = table_3.drop(['rankings_rank','id_1','idMal_1','site','type'], axis=1)
```

```python
table_3.dropna(subset="sites",inplace=True)
```

```python
table_3['external_sites'] = table_3['sites']
```

```python
table_3 = table_3.drop(['sites'], axis=1)
```

```python
def expand_column(column_name):
    # Check if the value is a list; if yes, return it as is
    if isinstance(column_name, list):
        return column_name
    # Otherwise, process it as a string
    return column_name.strip("[]").replace("'", "").split(", ")
```

```python
table_3['genres'] = table_3['genres'].apply(expand_column)
table_3['tags'] = table_3['tags'].apply(expand_column)
```

```python
table_3 = table_3.explode('genres').explode('tags')
```

```python
table_3.isnull().sum()
```
