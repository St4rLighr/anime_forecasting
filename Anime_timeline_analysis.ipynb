```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/Original Anime dataset  - anime-dataset-2023.csv')
df.head()
```

```python
df.drop(columns=['Start Air Date', 'End Date', 'Aired','English name'], inplace=True)
df.head()
```

```python
df.dtypes
```

```python
df.columns = df.columns.str.strip()
```

```python
df.rename(columns={'Start Date': 'Start_date', 'End Date.1': 'End_date', 'English name':'English_name','Scored By':'Scored_by'}, inplace=True)
df.head()
```

```python
df['Start_date'] = pd.to_datetime(df['Start_date'], format='%d/%m/%Y', errors='coerce')
df['End_date'] = pd.to_datetime(df['End_date'], format='%d/%m/%Y', errors='coerce')
df['Score'] = pd.to_numeric(df['Score'],errors='coerce')
df['Rank'] = pd.to_numeric(df['Rank'],errors='coerce')
df['Popularity'] = pd.to_numeric(df['Popularity'],errors='coerce')
df['Members'] = pd.to_numeric(df['Members'],errors='coerce')
df['Favorites'] = pd.to_numeric(df['Favorites'],errors='coerce')
df['Scored_by'] = pd.to_numeric(df['Scored_by'],errors='coerce')
```

```python
df.isnull().sum()
```

```python
filtered_df= df[df['Type'] == 'TV']   #filter data to TV only
filtered_df.head()
```

```python
csv_file_path = 'filtered_anime_data.csv'
filtered_df.to_csv(csv_file_path, index=False)
```

```python
filtered_df.shape
```

```python
filtered_df.isnull().sum()
```

```python
filtered_df= filtered_df.dropna(subset=['Start_date'])
filtered_df.isnull().sum()
```

```python
df_cleaned_sorted = filtered_df.sort_values(by='Start_date')
df_cleaned_sorted.head()


```

```python
df_cleaned_sorted.shape
```

```python
df_cleaned_sorted['Day'] = df_cleaned_sorted['Start_date'].dt.day
df_cleaned_sorted['Month'] = df_cleaned_sorted['Start_date'].dt.month
df_cleaned_sorted['Year'] = df_cleaned_sorted['Start_date'].dt.year
```

```python
print(df_cleaned_sorted[['Start_date', 'Day', 'Month', 'Year']].head())
```

```python
df_cleaned_sorted.head()
```

```python
csv_file_path = 'anime_dataset_including_2024_2025.csv'
df_cleaned_sorted.to_csv(csv_file_path, index=False)
```

```python
# # NUMBER OF ANIME RELEASED PER YEAR
# # group by year

# anime_count_per_year = df_cleaned_sorted.groupby('Year').size()
# print(anime_count_per_year)
```

```python
df_excluded_2024_2025 = df_cleaned_sorted[
    (df_cleaned_sorted['Start_date'].dt.year != 2024) &
    (df_cleaned_sorted['Start_date'].dt.year != 2025)
]
print(df_excluded_2024_2025 [['Name', 'Start_date']])
```

```python
anime_count_per_year = df_excluded_2024_2025.groupby('Year').size()
print(anime_count_per_year)
```

```python
anime_count_df = anime_count_per_year.reset_index(name='Anime Count')
```

```python
anime_count_df.columns = ['Year', 'Anime Count']
```

```python
csv_file_path = 'anime_count_per_year_excluding_2024_2025.csv'
anime_count_df.to_csv(csv_file_path, index=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
fig = px.bar(anime_count_df,
             x='Year',
             y='Anime Count',
             title='Number of Anime Released Per Year',
             labels={'Year': 'Year', 'Anime Count': 'Number of Anime'},
             color_discrete_sequence=['#1f77b4']
)

fig.update_layout(
    xaxis_title='Year',
    yaxis_title='Number of Anime',
    title_font_size=16,
    xaxis_tickfont_size=12,
    yaxis_tickfont_size=12,
)


fig.show()
```

```python
fig = px.line(anime_count_df, x='Year', y='Anime Count', markers=True,
              title='Number of Anime Released Per Year',
              labels={'Year': 'Year', 'Anime Count': 'Number of Anime'})


fig.show()
```
